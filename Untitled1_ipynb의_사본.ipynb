```
import requests
from bs4 import BeautifulSoup
```

```
page = requests.get("https://www.rottentomatoes.com/top/bestofrt/top_100_action__adventure_movies/")
soup = BeautifulSoup(page.content)
```

```
#@title 기본 제목 텍스트
print (soup.prettify())
```

```
thead = soup.find('thead')
ths = thead.find_all('th')
dates = [th.get_text() for th in ths]
dates
```

```
tbody = soup.find_all('tbody')
tbody = tbody[2]
trs = tbody.find_all('tr')
#trs = trs[0:]
trs
```

```
temp = trs[0]
nsd = [i.get_text() for i in temp.find_all('td')]
nsd
```

```
score = tbody.find_all('span',class_='tMeterScore')
scores = [sc.get_text() for sc in score]
scores
```

```
title=tbody.find_all('a',class_='unstyled articleLink')
titles = [t.get_text() for t in title]
titles
```

```
result = []
for tr in trs:
  tds = tr.find_all('td')
  temp = [i.get_text() for i in tds]
  result.append(temp)
result
```

```
#@title 기본 제목 텍스트
dict = {'score': scores, 'title':titles}
#for i in result:
#   dict[i[0]]= i[1:]
  
dict
```
